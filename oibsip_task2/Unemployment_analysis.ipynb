Importing Libraries and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the data
ud =pd.read_csv('Unemployment in India.csv')

ud.head()
```

Describe the dataset

```python
ud.describe()
```

Checking Missing Values 

```python

print("Missing values in the dataset: ")
print(ud.isnull().sum())
```

```python
print(ud["Region"].unique())
```

Mean analysis Of employment data at different region

```python
#region_stats = ud.groupby(['Region'])[['Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']].mean().round(2).reset_index()
region_stats = ud.groupby("Region").mean().reset_index()
print("\nMean of the columns for each region:\n")
region_stats
```

```python
plt.figure(figsize=(50, 20))
sns.set_style("whitegrid")
sns.barplot(x=region_stats['Region'], y=region_stats[' Estimated Unemployment Rate (%)'])
plt.title("Average Unemployment Rate by Region")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x=ud[' Estimated Employed'], y=ud[' Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate vs Employed Rate")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(ud.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Plot a boxplot of the unemployment rate by region
plt.figure(figsize=(40, 20))
sns.boxplot(x=ud['Region'], y=ud[' Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate by Region")
plt.show()
```

```python
# Plot a line chart of the unemployment rate over time
plt.figure(figsize=(25, 8))
sns.lineplot(x=ud[' Date'], y=ud[' Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate over Time")
plt.show()
```

```python
unemployment_by_state_and_region = ud[['Region', 'Area', ' Estimated Unemployment Rate (%)']].dropna()
fig = px.sunburst(unemployment_by_state_and_region, path=['Region','Area'], values=' Estimated Unemployment Rate (%)', title='Unemployment Rate by State and Region')
fig.update_traces(hovertemplate='<b>%{label} </b><br> %{percentRoot:.2%} of Total Unemployment Rate')
fig.show()
```
